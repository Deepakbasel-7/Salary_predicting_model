```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline
from matplotlib.pyplot import xticks
from matplotlib.pyplot import yticks
```

```python
dataset= pd.read_csv('Salary Data.csv')
dataset=dataset.dropna()
dataset
```

```python
dataset.info()
```

```python
gender= pd.get_dummies(dataset['Gender'], drop_first= True)
gender
```

```python
dataset= pd.concat([dataset, gender], axis= 1)
dataset
```

```python
dataset= dataset.drop('Gender',axis= 1)
dataset
```

```python
education= pd.get_dummies(dataset['Education Level'], drop_first=False)
education
```

```python
dataset= pd.concat([dataset, education], axis=1)
dataset
```

```python
dataset= dataset.drop(['Education Level','Job Title'], axis=1)
dataset
```

```python
dataset.corr()
```

```python
#Starting the machine Learning
import seaborn as sns
plt.figure(figsize=(16,9))
sns.heatmap(dataset.corr(), annot= True)
plt.show()
```

```python
sns.pairplot(dataset)
```

```python
dataset
```

```python
dataset= dataset.drop(['Male'], axis=1)
dataset
```

```python
dataset[['Salary','Years of Experience','Age']]= dataset[['Salary','Years of Experience','Age']].astype('int')
dataset
```

```python
from sklearn.model_selection import train_test_split
df_train, df_test= train_test_split(dataset, test_size= 0.15, random_state=1)
```

```python
df_train
```

```python
from sklearn.linear_model import LinearRegression
x_train= df_train[['Age','Years of Experience',"Bachelor's","Master's",'PhD']]
y_train= df_train['Salary']


#For linear Regression
lr= LinearRegression()
lr_model= lr.fit(x_train,y_train)

#For Logistic Regression
from sklearn.linear_model import LogisticRegression
lg= LogisticRegression()
lg_model= lg.fit(x_train, y_train)



#For Polynomial Regression
from sklearn.preprocessing import PolynomialFeatures
poly_reg= PolynomialFeatures(degree=8)
x_poly=poly_reg.fit_transform(x_train)
# y_poly= poly_reg.fit(y_train)
pol_reg= LinearRegression()
model= pol_reg.fit(x_poly, y_train)
poly_reg.fit(x_train, y_train)


```

```python
#To dump model
import pickle as pk
filename= 'model.pickle'
pk.dump(lr_model, open(filename, 'wb'))
```

```python
data= df_test.iloc[50:51]
actual_salary=data['Salary']
data= data.drop(['Salary'], axis= 1)
data
```

```python
#Linear model
print("Predicted Salary = ", lr_model.predict(data))
print("Actual Salary = ", actual_salary)

```

```python
#Logistic model
print("Predicted Salary = ", lg_model.predict(data))
print("Acutal Salary= ", actual_salary)

```

```python
#Polynomial regression model test

predicted_salary= pol_reg.predict(poly_reg.fit_transform(data))
print("Actual Salary = ", actual_salary)
print("Predicted Salary= ", predicted_salary)
```

```python
#Accuracy check of train data in all models

print("Linear Regression=", lr_model.score(x_train, y_train)*100)
print("Logistic Regression=", lg_model.score(x_train, y_train)*100)
print("Polynomial Regression=", pol_reg.score(poly_reg.fit_transform(x_train), y_train)*100)
```

```python
#Accuracy check of test data in all models
test_data= df_test
y_test= test_data['Salary']
x_test= test_data.drop(['Salary'], axis=1)

print("Linear Regression=", lr_model.score(x_test, y_test)*100)
print("Logistic Regression=", lg_model.score(x_test, y_test)*100)
print("Polynomial Regression=", pol_reg.score(poly_reg.fit_transform(x_test), y_test)*100)

```
